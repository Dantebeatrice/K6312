```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import math
from patsy import dmatrices


import statsmodels.discrete.discrete_model as sm
import statsmodels.formula.api as smf
from statsmodels.graphics.regressionplots import *
from sklearn import datasets, linear_model
from sklearn.metrics import confusion_matrix
```

```python
Default = pd.read_csv('data/Default.csv', header=0)
```

```python
Default.head()
```

```python
y, X = dmatrices('default~income+balance', Default, return_type = 'dataframe')
print (y)
```

```python
print(X)
```

## Question(a)

```python
logit = sm.Logit(y.iloc[:,1], X)
print(logit.fit().summary())
```

```python
logit.fit().predict()
```

```python
predict_label = pd.DataFrame(np.zeros(shape=(10000,1)), columns = ['label'])
```

```python
predict_label.iloc[logit.fit().predict()>0.5] = 1
# this works also: predict_label[logit.fit().predict()>0.5] = 1
```

```python
print(predict_label)
```

```python
confusion_matrix(y.iloc[:,1], predict_label.iloc[:,0])
```

```python
y_actu = pd.Series(y.iloc[:,1], name='Actual')
y_pred = pd.Series(predict_label.iloc[:,0], name='Predicted')
df_confusion = pd.crosstab(y_actu, y_pred)
print(df_confusion)
```

```python
np.mean(y.iloc[:,1] == predict_label.iloc[:,0]) # to get accuracy
```

## Question(b)

```python
Default_train = Default[Default.index < 7000]
Default_test = Default[Default.index >= 7000]
```

```python
Default_train
```

```python
Default_train.shape
```

```python
Default_test.shape
```

```python
y_train, X_train = dmatrices('default~income+balance', Default_train, return_type = 'dataframe')
y_test, X_test = dmatrices('default~income+balance', Default_test, return_type = 'dataframe')
```

```python
logit = sm.Logit(y_train.iloc[:,1], X_train)
print (logit.fit().summary())
```

```python
preds = logit.fit().predict(X_test)
predict_label = pd.DataFrame(np.zeros(shape=(X_test.shape[0],1)), columns = ['label'])
threshold = 0.5
predict_label.iloc[preds.values > threshold] = 1  
confusion_matrix(y_test.iloc[:,1], predict_label.iloc[:,0])
```

```python
1-np.mean(y_test.iloc[:,1].values==predict_label.iloc[:,0].values) # to get accuracy
```

## Question (d)

```python
# refer to https://stackoverflow.com/questions/39163354/evaluating-logistic-regression-with-cross-validation
from sklearn.linear_model import LogisticRegression
from sklearn import metrics, model_selection
from sklearn import datasets
```

```python
predicted = model_selection.cross_val_predict(LogisticRegression(), X, y.iloc[:,1], cv=10, method='predict')
print (1-metrics.accuracy_score(y.iloc[:,1], predicted))
print (metrics.classification_report(y.iloc[:,1], predicted))
```
